```python
import pandas as pd
import seaborn as sns
import numpy as np
import json
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
tips.day.value_counts()
```

```python
days = tips.day.unique()

sunday = tips[tips['day'] == 'Sun'].iloc[:, :10]
saturday = tips[tips['day'] == 'Sat'].iloc[:, :10]
thursday = tips[tips['day'] == 'Thur'].iloc[:, :10]
friday = tips[tips['day'] == 'Fri'].iloc[:, :10]
```

```python
def create_dataset(df, name, step):    
    record = {}
#     record['x'] = [[2, 3] for day in df['day']]
    record['x'] = step
    record['y'] = [[tip, tip] for tip in df['tip']]
    record['z'] = [[total_bill, total_bill] for total_bill in df['total_bill']]
    color = "rgb(31,31,255)"
    number_scale = np.arange(11) * .1
    record['colorscale'] = [[str(number), color] for number in number_scale]
    record['day'] = name
    
    return record
```

```python
days = [sunday, saturday, thursday, friday]
names = ['sunday', 'saturday', 'thursday', 'friday']
steps = [[2, 3], [4, 5], [6, 7], [8, 9]]
```

```python
data = []

for day, name, step in zip(days, names, steps):
    record = create_dataset(day, name, step)
    data.append(record)
```

```python
with open('data.json', 'w') as f:
    json.dump(data, f)
```
